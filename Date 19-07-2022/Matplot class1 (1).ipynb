```python
plt.bar(classes,class1_stu,width=0.2,align='edge')
plt.show()
```

```python
plt.bar(classes,class1_stu,width=0.2,align='edge',color='y')
plt.show()
```

```python
plt.bar(classes,class1_stu,width=0.2,align='edge',color='y',edgecolor='r')
plt.show()
```

```python
plt.bar(classes,class1_stu,width=0.2,align='edge',color='y',edgecolor='r',linewidth=4)
plt.show()
```

```python
plt.bar(classes,class1_stu,width=0.2,align='edge',color='y',edgecolor='r',linewidth=4,alpha=0.6,linestyle='--')
plt.show()
```

```python
plt.bar(classes,class1_stu,width=0.2,align='edge',color='y',edgecolor='r',linewidth=4,alpha=0.6,linestyle='--',label='class1_stu')
plt.legend()
plt.show()
```

```python
plt.bar(classes,class1_stu,width=0.2,align='edge',color='y',edgecolor='r',linewidth=4,alpha=0.6,linestyle='--',
        label='class1_stu',visible=False)
plt.legend()
plt.show()
```

```python
plt.bar(classes,class1_stu,width=0.2,align='edge',color='y',edgecolor='r',linewidth=4,alpha=0.6,linestyle='--',
        label='class1_stu')
plt.legend()
plt.title("Class_graph")
plt.xlabel("Classes")
plt.ylabel("Class1_stu")
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.bar(classes,class1_stu,width=0.2,align='edge',color='y',edgecolor='r',linewidth=4,alpha=0.6,linestyle='--',
        label='class1_stu')
plt.legend()
plt.title("Class_graph")
plt.xlabel("Classes")
plt.ylabel("Class1_stu")
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.bar(classes,class1_stu,width=0.2,color='y',linewidth=4,alpha=0.6,label='class1_stu')
plt.bar(classes,class2_stu,width=0.2,color='g',linewidth=4,alpha=0.6,label='class2_stu')
plt.legend()
plt.title("Class_graph")
plt.xlabel("Classes")
plt.ylabel("Class1_stu")
plt.show()
```

```python
plt.figure(figsize=(16,9))
class_index=np.arange(len(classes))
plt.bar(classes,class1_stu,color='y',label='class1_stu')
plt.bar(classes,class2_stu,color='g',label='class2_stu')
plt.legend()
plt.title("Class_graph")
plt.xlabel("Classes")
plt.ylabel("Class1_stu")
plt.show()
```

```python
plt.figure(figsize=(16,9))
class_index=np.arange(len(classes))
width=0.2
plt.bar(class_index,class1_stu,width,color='y',label='class1_stu')
plt.bar(class_index,class2_stu,width,color='g',label='class2_stu')
plt.legend()
plt.title("Class_graph")
plt.xlabel("Classes")
plt.ylabel("Class1_stu")
plt.show()
```

```python
plt.figure(figsize=(16,9))
class_index=np.arange(len(classes))
width=0.2
plt.bar(class_index,class1_stu,width,color='y',label='class1_stu')
plt.bar(class_index+width,class2_stu,width,color='g',label='class2_stu')
plt.legend()
plt.title("Class_graph")
plt.xlabel("Classes")
plt.ylabel("Class1_stu")
plt.show()
```

```python
plt.figure(figsize=(16,9))
class_index=np.arange(len(classes))
width=0.2
plt.bar(class_index,class1_stu,width,color='y',label='class1_stu')
plt.bar(class_index+width,class2_stu,width,color='g',label='class2_stu')
plt.bar(class_index+width+width,class3_stu,width,color='b',label='class3_stu')
plt.legend()
plt.title("Class_graph")
plt.xlabel("Classes")
plt.ylabel("Class1_stu")
plt.show()
```

```python
plt.figure(figsize=(16,9))
class_index=np.arange(len(classes))
width=0.2
plt.bar(class_index,class1_stu,width,color='y',label='class1_stu')
plt.bar(class_index+width,class2_stu,width,color='g',label='class2_stu')
plt.bar(class_index+width+width,class3_stu,width,color='b',label='class3_stu')
plt.xticks(class_index+width,classes)
plt.legend()
plt.title("Class_graph")
plt.xlabel("Classes")
plt.ylabel("Class1_stu")
plt.show()
```

# Histogram

```python
import random
```

```python
ml_students_age=np.random.randint(18,45,(100))
py_students_age=np.random.randint(18,450,(100))
```

```python
#total 100 random data from 18-45
```

```python
print(ml_students_age)
print(py_students_age)
```

```python
plt.hist(ml_students_age)
```

```python
plt.hist(ml_students_age)
plt.show()
```

```python
plt.hist(ml_students_age)
plt.title("ML data")
plt.show()
```

```python
plt.hist(ml_students_age)
plt.title("ML data")
plt.xlabel("Students_age")
plt.ylabel("No_of_students")
plt.show()
```

```python
#range or frequency of data is called bins
bins=[15,20,25,30,35,40,45]
plt.hist(ml_students_age,bins)
plt.title("ML data")
plt.xlabel("Students_age")
plt.ylabel("No_of_students")
plt.show()
```

```python
#rwidth=row width
bins=[15,20,25,30,35,40,45]
plt.hist(ml_students_age,bins,rwidth=0.8)
plt.title("ML data")
plt.xlabel("Students_age")
plt.ylabel("No_of_students")
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_students_age,bins,rwidth=0.8,histtype="step",align='right')
plt.title("ML data")
plt.xlabel("Students_age")
plt.ylabel("No_of_students")
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_students_age,bins,rwidth=0.8,histtype="bar",align='right',orientation='horizontal')
plt.title("ML data")
plt.xlabel("Students_age")
plt.ylabel("No_of_students")
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_students_age,bins,rwidth=0.8,histtype="bar",align='right',orientation='horizontal',color='b')
plt.title("ML data")
plt.xlabel("Students_age")
plt.ylabel("No_of_students")
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist(ml_students_age,bins,rwidth=0.8,histtype="bar",align='right',color='b',
         label="ml student")
plt.hist(py_students_age,bins,rwidth=0.8,histtype="bar",align='right',color='y',
         label="py student")
plt.title("ML data")
plt.xlabel("Students_age")
plt.ylabel("No_of_students")
plt.legend()
plt.show()
```

```python
bins=[15,20,25,30,35,40,45]
plt.hist([ml_students_age,py_students_age],bins,rwidth=0.8,histtype="bar",align='right',color=['b','y']
         ,label=["ml student","py_student"])
plt.title("ML data")
plt.xlabel("Students_age")
plt.ylabel("No_of_students")
plt.legend()
plt.show()
```

# PieChart

```python
slices=[7,2,2,13]
activities=['sleeping','eating','working','playing']
cols=["c","m","r","b"]
```

```python
plt.pie(slices)
plt.show()
```

```python
plt.pie(slices,labels=activities)
plt.show()
```

```python
plt.pie(slices,labels=activities,colors=cols)
plt.show()
```

```python
plt.pie(slices,labels=activities,colors=cols,startangle=90)
plt.show()
```

```python
plt.pie(slices,labels=activities,colors=cols,startangle=90,explode=(0,0,.2,0))
plt.show()
```

```python
plt.pie(slices,labels=activities,colors=cols,startangle=90,explode=(.2,0,.2,0))
plt.show()
```

```python
plt.pie(slices,labels=activities,colors=cols,startangle=90,explode=(.2,0,.2,0),autopct="%1.2f%%")
plt.show()
```

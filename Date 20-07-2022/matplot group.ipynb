```python
slices = [7,2,2,13]
activities = ['sleeping','eating','working','playing']
```

```python
cols=["c","m","r","b"]
```

```python
plt.pie(slices)
plt.show()
```

```python
plt.pie(slices,labels=activities)
plt.show()
```

```python
plt.pie(slices,labels=activities,colors=cols)
plt.show()
```

```python
plt.pie(slices,labels=activities,colors=cols,startangle=90,explode=(0,0,.2,0))
plt.show()
```

```python
plt.pie(slices,labels=activities,colors=cols,startangle=90,explode=(.1,0,.2,0))
plt.show()
```

```python
plt.pie(slices,labels=activities,colors=cols,startangle=90,explode=(.1,0,.2,0),shadow=True,autopct="%1.1f%%")
plt.show()
```
